```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customers Data.csv')
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df['email'] = df['email'].astype(str)
```

```python
df.loc[~df['email'].str.match(r"[^@]+@[^@]+\.[^@]+"), 'email'] = np.nan
```

```python
df['email'] = df['email'].fillna('abc@gmail.com')
```

```python
df['name'] = df['name'].fillna('Tejas')
```

```python
df['name'] = df['name'].fillna('Tejas')
```

```python
df.to_csv('employee_cleaned2.csv', index=False)

print("Data cleaning completed successfully.")
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df, y='city', order=df['city'].value_counts().index)

plt.title('Distribution of Customers by City')
plt.xlabel('Number of Customers')
plt.ylabel('City')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(data=df, x='age', bins=10, kde=True)

plt.title('Distribution of Customers by Age')
plt.xlabel('Age')
plt.ylabel('Number of Customers')

plt.show()
```

```python
df_sorted = df.sort_values(by='age').reset_index(drop=True)

plt.figure(figsize=(12,6))

sns.lineplot(data=df_sorted, x=df_sorted.index, y='age', marker='o')

plt.title('Sorted Distribution of Customers by Age')
plt.xlabel('Customer Index')
plt.ylabel('Age')

plt.grid(True)

plt.show()
```
