```python
import pandas as pd

data = {
    "Name": ["Rahul","Priya","Amit","Sneha","Karan","Neha","Arjun","Pooja"],
    "Age": [22,25,23,28,30,27,24,26],
    "Salary": [40000,50000,45000,60000,65000,55000,48000,52000],
    "Experience": [1,3,2,5,6,4,2,3],
    "Department": ["IT","HR","IT","Finance","Finance","HR","IT","Finance"]
}

df = pd.DataFrame(data)

df.to_csv("employee_dataset.csv", index=False)

df
```

```python
df = pd.read_csv("employee_dataset.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Salary"], bins=8)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="Department")

plt.title("Employees by Department")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x="Experience", y="Salary")

plt.title("Experience vs Salary")

plt.show()
```
